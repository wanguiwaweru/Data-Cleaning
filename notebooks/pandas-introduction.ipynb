Use df.set_index() to specify the column to be used as the index column.

# Subsetting Data

```python
df.shape
```

We can subset the dataframe by specifying the columns to be included in the dataframe and overwriting the current df.

```python
# Subsetting Columns
df = df[['title', 'description', 'publishTime',
         'duration_seconds', 'viewCount', 'likeCount',
         'commentCount']]
```

```python
# Subsetting using loc
df_subset1 = df.loc[df['viewCount'] > 1_000_000]
df_subset1.head(2)
```

df.query() returns the columns that meet the specified criteria in the query

> It offers a cleaner alternative to passing the boolean expression to df.loc[]

```python
df_subset2 = df.query('viewCount > 1000000')
df_subset2.head(2)
```

Use isna() function to get rows with null values. It returns booleans of whether row has NaN value or not

use ~ to negate the results of the query.

```python
# This returns the non-null entries because its negated using ~
df = df.loc[~df['likeCount'].isna()]
df.tail(3)
```

# Casting dtypes

When data is read into a pandas dataframe, pandas converts each column into a given datatype depending on the values stored in the series.

For instance,datetimes can be set to strings and any numeric column that has a Null is set to float. This can be cast to int when the null values are removed.

Casting into different data types reduces the memory used to store the data and can increase speed of reading and writing the data  

```python
#cast to int since the null values are removed
df['viewCount'] = df['viewCount'].astype('int')
df['likeCount'] = df['likeCount'].astype('int')
```

convert to date time using pd.to_datetime()
convert to numeric type using to_numeric(). Pandas can convert the column to type int or float.

```python
df['publishTime'] = pd.to_datetime(df['publishTime'])
```

```python
df['likeCount'] = pd.to_numeric(df['likeCount'].astype('str'))
```
