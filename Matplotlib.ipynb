```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Assignment_Solutions/master/Standard%20Metropolitan%20Areas%20Data%20-%20train_data%20-%20data.csv")
data.head()
```

Plot a histogram of one column 

```python
plt.hist(data.graduates)
plt.show()
```

```python
plt.bar(data.region, data.crime_rate, color='cyan')
plt.show()
```

```python
plt.barh(data.region, data.crime_rate, color='maroon')
plt.show()
```

```python
plt.figure(figsize=(10,5))
browsers = ['chrome', 'mozilla','IE', 'Edge', 'bravo']
market_share = [45,20,5,15,15]

Explode = [0.1,0,0,0,0]
plt.pie(market_share,explode=Explode,labels=browsers,autopct='%1.1f%%')
plt.show()
```

```python
plt.plot([2,4,6],[3,5,7,])
plt.show()
```

```python
even = np.array([2,4,6,8,10])
odd = np.array([3,5,7,9,11])
plt.plot(even,odd)
plt.title('Odd Even')
plt.xlabel('Even')
plt.ylabel('Odd')
plt.figure(figsize =(25,5))
plt.show()
```

```python
plt.scatter(data.percent_senior, data.income,color = 'cyan')
plt.show()
```

```python
plt.plot(data.physicians, data.income,color = 'cyan', label = 'physicians')
plt.plot(data.work_force, data.income,color = 'red', label = 'workforce')
plt.legend()
plt.show()
```

```python
plt.plot(data.physicians, data.income,marker = 'o', linestyle = '--') 
plt.show()
```

```python
plt.plot(data.physicians, data.income, 'bo--') 
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data.work_force, data.income,color = 'cyan')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(data.physicians, data.income,'b^')
plt.title('physicians v income')
plt.subplot(1,2,2)
plt.plot(data.work_force, data.income,'g^')
plt.title('workforce v income')
plt.suptitle('Multiple plots')
plt.show()
```
